```python
import pandas as pd
import numpy as np
#Crear el diccionario
data = {
'Producto':['Laptop', 'Mouse', 'Monitor', 'Teclado'],
'Precio': [750000, 20000, 50000, 35000],
'Stock':[10, 20, 15, 25]
}
#Transformar a dataframe
print("\n" + "*"*30 + "\n")
df = pd.DataFrame(data)
#Mostrar dataframe
print(df)
print("*"*30 + "\n")
#Mostrar el promedio de stock de los productos.
prom_stock = df["Stock"].mean()
print(f"El promedio de Stock de productos es de {prom_stock:.0f}")
print("\n" + "*"*30 + "\n")

#Filtrar los productos con un stock igual o mayor a 20
prod_20 = df[df["Stock"] >=20]
print(prod_20)
print("\n" + "*"*30 + "\n")

#Mostrar el promedio del precio
precio_prom=df["Precio"].mean()
print(f"El precio promedio de los productos es de: {precio_prom:.0f}")
print("\n" + "*"*30 + "\n")

#Mostrar la cantidad total de todos los productos
total=df["Stock"].sum()
print(f"La cantidad total de todos los productos es de :{total}")
print("\n" + "*"*30 + "\n")

#Mostrar la información de los productos con precio mayor o igual a 500.000
precio_500000 = df[df["Precio"] >=500000]
print(f"Los productos con precio mayor o igual a 500.000 son: \n{precio_500000}")
```
